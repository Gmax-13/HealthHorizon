```python
import matplotlib.pyplot as plt

# Sample data: Days of the week and corresponding calorie intake
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
calories = [2200, 2100, 2300, 2500, 2000, 2700, 2400]

# Plot the graph
plt.figure(figsize=(8, 5), frameon=False)
plt.plot(days, calories, marker='o', linestyle='-', color='g')

# Show only x-axis labels (days)
plt.xticks(days)
plt.yticks([])

# Remove borders
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["left"].set_visible(False)
plt.gca().spines["bottom"].set_visible(False)

# Show the graph
plt.show()
```
